```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import explained_variance_score
```

```python
datos = pd.read_csv("tic-tac-toe.data.txt", header = None)
```

```python
datos.head()
```

```python
datos.dtypes
```

```python
funcion = preprocessing.LabelEncoder()
for f in datos:
    datos[f] = funcion.fit_transform(datos[f].astype(str))  
datos.head()
```

```python
y = datos.pop(9)
```

```python
modelos = []
modelos.append(("Lineal",LinearRegression()))
modelos.append(("Ridge", Ridge()))
modelos.append(("Lasso",Lasso()))
modelos.append(("ElasticNet",ElasticNet()))
modelos.append(("Forest",RandomForestRegressor()))
modelos.append(("Baggin",BaggingRegressor()))
modelos.append(("extree",ExtraTreesRegressor()))
modelos.append(("arbol",DecisionTreeRegressor()))
modelos.append(("SVR",SVR()))
modelos.append(("Knn",KNeighborsRegressor()))


for nombre, modelo in modelos:
    
    modelo.fit(datos,y)
    prediccion = modelo.predict(datos)
    score = explained_variance_score(y,prediccion)
    error = mean_absolute_error(prediccion, y)
    
    print("modelo %s: %f , error: %f"% (nombre,score,error))
```
